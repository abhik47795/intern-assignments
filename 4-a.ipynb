```python
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
```

```python
page=requests.get(url)
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
youtube_data=soup.find_all('table')[0]
```

```python
import pandas as pd
```

```python
#creating a data frame and storing the table
```

```python
df=pd.read_html(str(youtube_data))[0]
```

```python
df
```

```python
df.drop(columns=['Note','Unnamed: 6'],axis=1,inplace=True)
```

```python
df
```

```python
#saving the data in csv form
```

```python
df.to_csv('most_viewed.csv')
```
