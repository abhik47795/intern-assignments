```python
import bs4
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url='http://statisticstimes.com/economy/gdp-of-indian-states.php'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
#scrapping the table at index 1 using bs4 
```

```python
data=soup.find_all('table')[1]
```

```python
import pandas as pd
```

```python
df=pd.read_html(str(data))[0]
```

```python
df.drop('GSDP (Cr INR at 2011-12 prices)',axis=1,inplace=True)
import warnings
warnings.filterwarnings('ignore')
```

```python
df
```

```python
df.to_csv('gdp_state_wise.csv')  #saving the file in csv format
```
