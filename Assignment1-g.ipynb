# Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating

```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.amazon.in/s?k=smartphone+under+20000&crid=1XQVORY843OSA&sprefix=smartpho%2Caps%2C670&ref=nb_sb_noss_1'
```

```python
page=requests.get(url)
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
name=soup.find_all('h2',class_='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
```

```python
name[0:4]
```

```python
names=[]
```

```python
for i in name:
    names.append(i.get_text())
```

```python
names[0:4]
```

```python
price=[]
image=[]
rating=[]
```

```python
rate=soup.find_all('span',class_='a-price-whole')
```

```python
rate[0:4]
```

```python
for i in rate:
    price.append(i.get_text())
```

```python
price[0:5]
```

```python
review=soup.find_all('div',class_='a-row a-size-small')
```

```python
review[0:4]
```

```python
for i in review:
    rating.append(i.get_text())
```

```python
rating[0:4]
```

```python
import re
images = soup.find_all('img', {'src':re.compile('.jpg')})
```

```python
images
```

```python
for i in images: 
    image.append(i.get_text())
```

```python
image=list(image)
```

```python
type(image)
```

```python
image=image[0:16]
```

```python
import pandas as pd
```

```python
amazon=pd.DataFrame({})
```

```python
print(len(image),len(rating),len(name),len(price))
```

```python
amazon['Product Name']=name
amazon['Price']=price
amazon['Image URL']=image
amazon['Ratings']=rating
```
