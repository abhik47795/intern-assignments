# Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://bookpage.com/reviews'
```

```python
page=requests.get(url)
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
book_name=[]
author_name=[]
genre=[]
review=[]
```

```python
name=soup.find_all('h4',class_='italic')
```

```python
name[0:9]
```

```python
for i in name:
    book_name.append(i.get_text())
```

```python
book_name=book_name[0:5]
book_name
```

```python
author=soup.find_all('p',class_='sans bold')
```

```python
author[0:10]
```

```python
for i in author:
    author_name.append(i.get_text())
```

```python
author_name=author_name[0:5]
author_name
```

```python
category=soup.find_all('p',class_='genre-links hidden-phone')
```

```python
category[0:5]
```

```python
for i in category:
    genre.append(i.get_text())
```

```python
genre=genre[0:5]
genre
```

```python
reviews=soup.find_all('div',class_='flex-article-content')
```

```python
reviews[0:5]
```

```python
#here i have scrapped the full content...
```

```python
for i in reviews:
    review.append(i.get_text())
```

```python
review=review[0:5]
review
```

```python
print(len(book_name),len(author_name),len(genre),len(review))
```

```python
import pandas as pd
```

```python
book_page=pd.DataFrame({})
```

```python
book_page['book name']=book_name
book_page['author name']=author_name
book_page['genre']=genre
book_page['review']=review
```
